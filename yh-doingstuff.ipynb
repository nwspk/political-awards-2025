```python
import pandas as pd
scores_df = pd.read_csv('data/Grantmaking Scoring Sheet (Human Powered) - Projects.csv')
```

```python
scores_df = scores_df.drop(columns=['Unnamed: 0', 'Unnamed: 19', 'Unnamed: 20', 'Reviewer',
                                    'Notes', 'Do I think it is cool?', 'Do I have higher level of knowledge about it?'])
scores_df
```

```python
scores_df['Problem type (qualitative)'].unique()[:5]
```

```python
import re
def clean_text(text):
    # Find all the text before '–'
    matches = re.findall(r'(^|,\s*)\"?([^–\"]+)\s*–', text)
    # Clean up whitespace and quotation marks, if any
    clean_matches = [match[1].strip().replace('"', '').lower() for match in matches]
    # Join the cleaned matches with commas
    return ','.join(clean_matches)
```

```python
scores_df['problem_type'] = scores_df['Problem type (qualitative)'].apply(clean_text)
scores_df['problem_type']
```

```python
# Split the 'problem_type' column into dummy variables
dummy_columns = scores_df['problem_type'].str.get_dummies(sep=',')

# Concatenate the dummy columns back to the original DataFrame
scores_df = pd.concat([scores_df, dummy_columns], axis=1)

scores_df = scores_df.rename(columns={'other': 'other problem type'})

scores_df
```

```python
scores_df['Intervention methodology'].unique()[:10]
```

```python
def clean_method(text):
    # Find all the text between """ and :
    matches = re.findall(r'"""([^:]+):', text)
    
    if 'Other' in text:
        matches.append('Other')

    # Clean up whitespace and quotation marks, if any
    clean_matches = [match.replace('"', "").replace(',', '').strip().lower() for match in matches]
    # Join the cleaned matches with commas
    return ','.join(clean_matches)
```

```python
scores_df['intervention_methodology'] = scores_df['Intervention methodology'].apply(clean_method)
scores_df['intervention_methodology'].unique()[:10]
```

```python
# Split the 'problem_type' column into dummy variables
dummy_columns = scores_df['intervention_methodology'].str.get_dummies(sep=',')

# Concatenate the dummy columns back to the original DataFrame
scores_df = pd.concat([scores_df, dummy_columns], axis=1)

scores_df = scores_df.rename(columns={'other': 'other intervention methodology'})

scores_df
```

```python
# Ensure each combination is treated as a tuple (or frozenset for unordered uniqueness)
scores_df["prob_meth_combined"] = scores_df.apply(lambda row: (tuple(sorted(row["problem_type"])), tuple(sorted(row["intervention_methodology"]))), axis=1)

# Count occurrences of each unique combination
combination_counts = scores_df["prob_meth_combined"].value_counts().reset_index()
combination_counts.columns = ["Combination", "Count"]
combination_counts
```

```python
top10_combo = combination_counts.nlargest(10, 'Count')
top10_combo
```

```python
import numpy as np
scores_df['grouping'] = np.nan

for index, row in top10_combo.iterrows():
    for i, r in scores_df.iterrows():
        if pd.notna(r['grouping']):
            continue
        else:
            if all(set(row['Combination'][t]).issubset(r['prob_meth_combined'][t]) for t in range(len(row['Combination']))):
                scores_df.loc[i, 'grouping'] = index
            else:
                pass
```

## Reduce scoring dimensions

### Step 1. Normalize/standardize scores

```python
merged_df[scoring_dimensions]
```

```python
from sklearn.preprocessing import StandardScaler
scores = merged_df[scoring_dimensions].values
scores = StandardScaler().fit_transform(scores)
scores
```

```python
normalized_scores = pd.DataFrame(scores, columns=scoring_dimensions)
normalized_scores = normalized_scores.merge(merged_df[['Project Name', 'grouping']], left_index=True, right_index=True)
normalized_scores
```

### Step 2. Use PCA to reduce dimensions to 2

```python
from sklearn.decomposition import PCA
pca_scores = PCA(n_components=2)
principalComponents_scores = pca_scores.fit_transform(normalized_scores[scoring_dimensions])
principalComponents_df = pd.DataFrame(
    principalComponents_scores,
    columns=['Principal Component 1', 'Principal Component 2'],
    index=normalized_scores.index  # Use the original index from normalized_scores
)
```

```python
principalComponents_scores
```

```python
final_score_df = pd.merge(normalized_scores, principalComponents_df, how='left', left_index=True, right_index=True)
final_score_df
```

```python
import matplotlib.pyplot as plt
import mplcursors
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Define a visually distinct colormap
unique_groups = final_score_df['grouping'].unique()
cmap = cm.get_cmap('tab10', len(unique_groups))  # Use 'tab10' for distinct colors
norm = mcolors.BoundaryNorm(boundaries=range(len(unique_groups) + 1), ncolors=len(unique_groups))

plt.figure(figsize=(12, 6))
scatter = plt.scatter(
    final_score_df['Principal Component 1'], 
    final_score_df['Principal Component 2'],
    c=final_score_df['grouping'].map({group: i for i, group in enumerate(unique_groups)}),  # Map groups to numeric values
    cmap=cmap,
    norm=norm,
    s=50  # Adjust point size
)

# Add a legend
handles = [plt.Line2D([0], [0], marker='o', color=cmap(i), linestyle='', markersize=8) for i in range(len(unique_groups))]
labels = [str(group) for group in unique_groups]
plt.legend(handles, labels, title='Grouping', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Scoring Dimensions')
plt.grid()

# Save the plot
plt.savefig('pca_scoring_dimensions_with_legend.png', bbox_inches='tight')
plt.show()
```

### Step 3. Do PCA again, but we reduce the dimensions separately for each group to 1

```python
final_score_copy = final_score_df.copy()
final_score_copy = final_score_copy.drop(columns=['Principal Component 1', 'Principal Component 2'])
```

```python
final_score_copy['grouping'].unique()
```

```python
final_score_by_group = []
for i in final_score_copy['grouping'].unique():
    group = final_score_copy[final_score_copy['grouping'] == i]
    pca_scores = PCA(n_components=1)
    principalComponents_scores = pca_scores.fit_transform(group[scoring_dimensions])
    principalComponents_series = pd.Series(principalComponents_scores.flatten(), index=group.index, name='Principal Component 1')
    group = group.join(principalComponents_series)
    group = group.drop(columns=scoring_dimensions)
    group = group.sort_values(by='Principal Component 1', ascending=False)
    final_score_by_group.append(group)
final_score_by_group[0].head(10)
```
