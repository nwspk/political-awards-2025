```python
import pandas as pd
scores_df = pd.read_csv('data/Grantmaking Scoring Sheet (Human Powered) - Projects.csv')
```

```python
scores_df = scores_df.drop(columns=['Unnamed: 0', 'Unnamed: 19', 'Unnamed: 20', 'Reviewer',
                                    'Notes', 'Do I think it is cool?', 'Do I have higher level of knowledge about it?'])
scores_df
```

```python
scores_df['Problem type (qualitative)'].unique()[:5]
```

```python
import re
def clean_text(text):
    # Find all the text before '–'
    matches = re.findall(r'(^|,\s*)\"?([^–\"]+)\s*–', text)
    # Clean up whitespace and quotation marks, if any
    clean_matches = [match[1].strip().replace('"', '').lower() for match in matches]
    # Join the cleaned matches with commas
    return ','.join(clean_matches)
```

```python
scores_df['problem_type'] = scores_df['Problem type (qualitative)'].apply(clean_text)
scores_df['problem_type']
```

```python
# Split the 'problem_type' column into dummy variables
dummy_columns = scores_df['problem_type'].str.get_dummies(sep=',')

# Concatenate the dummy columns back to the original DataFrame
scores_df = pd.concat([scores_df, dummy_columns], axis=1)

scores_df
```

```python
scores_df.columns
```

```python
scores_df['Intervention methodology'].unique()[:10]
```

```python
def clean_method(text):
    # Find all the text between """ and :
    matches = re.findall(r'"""([^:]+):', text)
    
    if 'Other' in text:
        matches.append('Other')

    # Clean up whitespace and quotation marks, if any
    clean_matches = [match.replace('"', "").replace(',', '').strip().lower() for match in matches]
    # Join the cleaned matches with commas
    return ','.join(clean_matches)
```

```python
scores_df['intervention_methodology'] = scores_df['Intervention methodology'].apply(clean_method)
scores_df['intervention_methodology'].unique()[:10]
```

```python
# Split the 'problem_type' column into dummy variables
dummy_columns = scores_df['intervention_methodology'].str.get_dummies(sep=',')

# Concatenate the dummy columns back to the original DataFrame
scores_df = pd.concat([scores_df, dummy_columns], axis=1)

scores_df
```

```python
scores_df = scores_df.drop(columns=['Intervention methodology', 'Problem type (qualitative)'])
scores_df.columns
```
